```python
import numpy as np
import pandas as pd
import pdb
```

```python
footballDataSet=pd.read_csv("Laliga.csv")
football=pd.DataFrame(footballDataSet)
football.head(50)
```

```python
#replace '-' with 0
football=football.replace('-', 0)
```

```python
#teams that exists between 1930 to 1980
#football['Debut']=pd.to_numeric(football['Debut'].str[:4])
teams = football[(football['Debut'] > 1930) & (football['Debut'] < 1980)]
teams
```

```python
#top five teams wrt points
football['Points']=football['Points'].astype(int)
football.nlargest(5, 'Points')
```

```python
def Goal_diff_count(typeValue):
    goalsDifferences = football.groupby('Team')['Team','GoalsFor','GoalsAgainst'].sum()
    goalsDifferences['GoalDifference'] = pd.to_numeric(goalsDifferences['GoalsFor'], errors='coerce') - pd.to_numeric(goalsDifferences['GoalsAgainst'], errors='coerce')
    if(typeValue =='difference'):        
        return goalsDifferences
    elif(typeValue=='maxmin'):
        maximumValue=goalsDifferences[['Team','GoalDifference']].max()
        minimumValue=goalsDifferences[['Team','GoalDifference']].min()
        return [maximumValue, minimumValue]
        
```

```python
#find goal differnce
teamGoalDifference=Goal_diff_count('difference')
teamGoalDifference
```

```python
#max and min value
maxandminValue=Goal_diff_count('maxmin')
print(maxandminValue)
```

```python
#winning percent
football['Winning Percent']= (pd.to_numeric(football['GamesWon'], errors='coerce')/pd.to_numeric(football['GamesPlayed'], errors='coerce'))*100
football['Winning Percent'].fillna(0, inplace=True)
football
```

```python
#top 5 winning percent teams
football['Winning Percent']=football['Winning Percent'].astype(str)
football["Winning Percent"].rank(method='max') 
football[['Team', 'Winning Percent']].head(5)
```
